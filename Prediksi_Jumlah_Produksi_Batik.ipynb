```python
import pandas as pd
pd.set_option('display.max_colwidth', None)
data = pd.read_excel('https://github.com/ifan007/Prediksi-Jumlah-Produksi-Batik/blob/main/just%20data%20batik%20campuran.xlsx?raw=true')
data
```

```python
data = data[['Kain/m','Malan/balok','Pewarna/sachet','Harga Jual/Rp', 'Target-Jumlah Produksi']]
data
```

```python
data.describe()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

plt.figure(figsize=(20,15))
sns.heatmap(data.corr(), annot=True)
plt.title("Heatmap", fontsize = 25);
plt.show()
```

```python
data.isnull().sum()
```

```python
data.isnull()
```

```python
data['Harga Jual/Rp'].fillna(data['Harga Jual/Rp'].mean(), inplace = True)
data['Target-Jumlah Produksi'].fillna(data['Target-Jumlah Produksi'].mean(), inplace = True)
```

```python
data.isnull().sum()
```

```python
x = data.iloc[:,:4].values
y = data.iloc[:,-1].values
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaler.fit(data)
x = scaler.fit_transform(x)
y = scaler.fit_transform(y.reshape(-1, 1))
```

```python
x
```

```python
y
```

```python
x.shape
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

# Create the neural network model
model = Sequential()
model.add(Dense(128, activation='relu', input_shape=(4,)))
model.add(Dense(64, activation='relu'))
model.add(Dense(1))
```

```python
from sklearn.metrics import mean_squared_error
# Compile the model
model.compile(optimizer='adam', loss='mean_squared_error')
```

```python
print(model.summary())
```

```python
# Train the model
history = model.fit(x_train, y_train, epochs=100, batch_size=5, verbose=1)
```

```python
# Evaluate the model
train_loss = model.evaluate(x_train, y_train, verbose=0)
test_loss = model.evaluate(x_test, y_test, verbose=0)
print("Train Loss:", train_loss)
print("Test Loss:", test_loss)
```

```python
# Make predictions
y_train_pred = model.predict(x_train)
y_test_pred = model.predict(x_test)
```

```python
# Inverse transform the scaled predictions
y_train_pred = scaler.inverse_transform(y_train_pred)
y_test_pred = scaler.inverse_transform(y_test_pred)
y_train = scaler.inverse_transform(y_train)
y_test = scaler.inverse_transform(y_test)
```

```python
# Calculate root mean squared error
import numpy as np
train_rmse = np.sqrt(mean_squared_error(y_train, y_train_pred))
test_rmse = np.sqrt(mean_squared_error(y_test, y_test_pred))
print("Train RMSE:", train_rmse)
print("Test RMSE:", test_rmse)
```

```python
# Visualize the results
plt.figure(figsize=(15, 10))
plt.plot(y_train, color='b', label='Actual')
plt.plot(y_train_pred, color='r', label='Predicted')
plt.title("Train Set - Actual vs Predicted")
plt.ylim(ymin=0,ymax=50)
plt.legend()
plt.show()
```

```python
# Visualize the results
plt.figure(figsize=(15, 10))
plt.plot(y_test, color='b', label='Actual')
plt.plot(y_test_pred, color='r', label='Predicted')
plt.title("Test Set - Actual vs Predicted")
plt.legend()
plt.show()
```
